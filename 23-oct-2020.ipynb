## Data Visualisation
- https://matplotlib.org/

```python
# Plot linear graph
import matplotlib.pyplot as plt
plt.title("Basic Linear Graph")
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
distance=list(range(2,11,2)) # 2,4,6,8,10
time=list(range(1,6)) # 1,2,3,4,5
plt.title('Distance vs Time')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(time,distance,'go')
plt.grid()
plt.show()
```

```python
# v=a*t
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i] * i for i in range(len(t))]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
# Voltage vs Time
import numpy as np
t=np.arange(0.1,2.0,0.1)
v=[1+np.sin(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.grid()
plt.xlabel("Time")
plt.ylabel("voltage")
plt.title("Time Vs Voltage")
plt.show()
```
